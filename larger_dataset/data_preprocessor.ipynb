```python
def get_all_file_names(base_name,limit_num):
    return [base_name + str(num) for num in range(1,limit_num + 1)]
```

```python
def read_chunks(files,path):
    #reads chunks csvs and converts them to a dataframe format
    final_df = []
    for file in files:
        df = pd.read_csv(path + file + ".csv")
        final_df.append(df)
        #function returns a list of dfs, or one df if only one input is given
    return final_df
```

```python
all_file_names = get_all_file_names("chunk_",60)
```

```python
def group_up_chunks(dfs):
    #adds up all dataframes together
    return pd.concat(dfs)
```

```python
for chunk in range(len(all_chunks)):
    preprocessed_text = []
    for i,text in enumerate(all_chunks[chunk]['text']):
        preprocessed_text.append(process_corpus(remove_stop_words(text)))
    all_chunks[chunk]['text'] = preprocessed_text
    all_chunks[chunk].to_csv(filepath + str(chunk + 1) + '.csv',index=False)
    print("progress: " + str(chunk + 1) + " out of " + str(len(all_chunks)))
```
