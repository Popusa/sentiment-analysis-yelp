```python
import pandas as pd
```

```python
df = pd.read_csv("yelp_review.csv")
```

```python
df
```

```python
def create_chunk(data,start_size,end_size):
    chunk = data.iloc[start_size:end_size]
    return chunk
```

```python
data_chunks = []
smaller_chunks_count = 9
larger_chunks_count = 6
LARGER_CHUNK_SIZE = 1000000
SMALLER_CHUNK_SIZE = 625000
LAST_CHUNK_SIZE = 261668
start_slice = 0
chunk_slice = SMALLER_CHUNK_SIZE

for i in range(smaller_chunks_count):
    if i == smaller_chunks_count:
        chunk_slice = start_slice + LAST_CHUNK_SIZE
        data_chunks.append(create_chunk(df,start_slice,LAST_CHUNK_SIZE))
    else:
        data_chunks.append(create_chunk(df,start_slice,chunk_slice))
        start_slice += SMALLER_CHUNK_SIZE
        chunk_slice += SMALLER_CHUNK_SIZE
```

```python
dataset_total = 0
for i in range(smaller_chunks_count):
    dataset_total += data_chunks[i].shape[0]
    print(data_chunks[i].shape)

print(dataset_total, dataset_total == df.shape[0])
```

```python
filepath = 'smaller_chunks/chunk_'
```

```python
for itr,chunk in enumerate(data_chunks):
    chunk.to_csv(filepath + str(itr + 1) + '.csv',index=False)
```
